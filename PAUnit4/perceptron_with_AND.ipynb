# Notebook overview

- Purpose: demonstrate a simple perceptron implementation of the AND logic gate.
- Key dependencies: numpy and a helper `step` function imported from `utils`.
- Structure:
    - Cell 1: imports (`numpy`, `step`).
    - Cell 2: defines `AND(x1, x2)` using weights, bias, `np.dot`, and `step`.
    - Cell 3: runs and prints the AND gate truth table for inputs (0,1).
- Expected output: a printed header "AND Gate" followed by the four input pairs and their corresponding outputs (0 or 1).

```python
import numpy as np
from utils import step
```

```python
def AND(x1, x2):
    weights = np.array([1,1])
    bias = -1
    output = np.dot([x1, x2], weights) + bias
    
    return step(output)
```

```python
print('AND Gate')
for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1, x2, '->' , AND(x1, x2))
```
