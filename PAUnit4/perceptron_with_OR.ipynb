# Neural Network Logic Gates

This notebook implements basic logic gates using a simple perceptron model with step activation function. It demonstrates how neural networks can learn fundamental logical operations through weighted inputs and bias terms.

## Contents

- **Utilities**: Uses a `step` function from a custom utils module for activation
- **OR Gate Implementation**: Defines an OR logic gate using weights and bias
- **Testing**: Validates the OR gate against all possible input combinations (0,0), (0,1), (1,0), (1,1)

The OR gate returns 1 when at least one input is 1, and 0 only when both inputs are 0.

```python
import numpy as np
from utils import step
```

```python
def OR(x1, x2):
    weights = np.array([1,1])
    bias = -0.5
    output = np.dot([x1, x2], weights) + bias
    
    return step(output)
```

```python
print("OR Gate")
for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1, x2, '->', OR(x1,x2))
```
